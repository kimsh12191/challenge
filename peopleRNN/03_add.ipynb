# 메뉴만들기

- 일단 lunch_kor 부터 진행한다.
    - menu split, 불러와서, 그메뉴들 count 랑 material받아옴.
    - 그 후에 true와 pred사이의 차이와의 관계? 찾기??..?
- 그다음 lunch_west 진행.


```python
menuKor = pd.merge(pred_submit_df[['date', 'true_kor', 'pred_kor', 'week']].set_index('date') ,
         menuDfraw.loc[menuDfraw.type2 == 'lunch_kor'][['date', 'menuSplit']].set_index('date'),
        left_index=True, right_index=True)
menuWest = pd.merge(pred_submit_df[['date', 'true_west', 'pred_west', 'week']].set_index('date') ,
         menuDfraw.loc[menuDfraw.type2 == 'lunch_west'][['date', 'menuSplit']].set_index('date'),
        left_index=True, right_index=True)
```

- kor menu 리스트 true, pred옆에 넣음

```python
menuKor.head()
```

- menuKor의 menuSplit에서 menuDfadd의 menu에 해당하는거 찾은 다음 menuMaterial로 바꿔줘야함

```python
error = 0
menuWest['menuMeterial'] = 0
for date, menuEachDay in enumerate(menuKor['menuSplit']):
    MeterialEachDay = []
    for eachMenu in menuEachDay:
        try:
            MeterialEachDay.append(menuDfadd[['menu', 'menuMaterial(육류, 해산물, 채소, 곡류 등)']].\
                loc[menuDfadd.menu==eachMenu]['menuMaterial(육류, 해산물, 채소, 곡류 등)'].values[0])
        except:
            error += 1
            print ('error index : {}'.format(date))
            print ('error menu : {}'.format(eachMenu))
            print ('# of error : {}'.format(error))
            MeterialEachDay.append('오류')
#     sys.stdout.write('\r'+str(date))
    menuWest['menuMeterial'].iloc[date] = MeterialEachDay
```
